```python
import pandas as pd

# Load the complete gear list
complete_list = pd.read_csv(r"C:\Users\TyHow\Desktop\complete.csv")

# Load individual activity lists
activity_list1 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\Backcountry Ski 6fe723ce44ff4de4bfdd6f14be6e5ca1_all.csv"
)
activity_list2 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\Backpacking List 956c6923c1c7401781f1116512d9d631_all.csv"
)
activity_list3 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\Ice Mixed Crag f769eb1f4fde4e9aab2ba5dfa42e1f57_all.csv"
)
activity_list4 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\International Comp 5ebc033e7eb942d79f5cf006c0a46f4a_all.csv"
)
activity_list5 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\Short Trip Home 0de4d1be650947849c6decc6fe86ae8d_all.csv"
)
activity_list6 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\weekend climb.csv"
)
activity_list7 = pd.read_csv(
    r"C:\Users\TyHow\Desktop\weekend winter.csv"
)

complete_list
```

```python
# Combine all activity lists
combined_activity_list = pd.concat(
    [
        activity_list1,
        activity_list2,
        activity_list3,
        activity_list4,
        activity_list5,
        activity_list6,
        activity_list7,
    ],
    ignore_index=True,
)

# Ensure the column names match for comparison (assuming 'Item' is the column name)
if isinstance(combined_activity_list, pd.DataFrame):
    combined_activity_list = combined_activity_list.iloc[:, 0].drop_duplicates()

if isinstance(complete_list, pd.DataFrame):
    complete_list = complete_list.iloc[:, 0]

# Identify missing items
missing_items = combined_activity_list[~combined_activity_list.isin(complete_list)]

# Print missing items to the console
print("Missing items:")
print(missing_items)

# Save the missing items to a new CSV file
missing_items.to_csv(r"C:\Users\TyHow\Desktop\missing_items.csv", index=False)

print("Missing items saved to 'missing_items.csv'")
```
