```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
csv_path = "/Users/thowe/Pictures/full_library_photo_metadata.csv" # Update this path if needed
df = pd.read_csv(csv_path)
print("Initial Data Shape:", df.shape)
```

```python
# Clean and convert numeric fields, assuming all columns should be numeric
df["ISO"] = pd.to_numeric(df["ISO"], errors="coerce")
df["FocalLength"] = pd.to_numeric(df["FocalLength"], errors="coerce")
df["FNumber"] = pd.to_numeric(df["FNumber"], errors="coerce")
df["ShutterSpeed"] = pd.to_numeric(df["ShutterSpeed"], errors="coerce")

# Leave missing values as NaN (no filling)
print("Final Data Shape (including NaNs):", df.shape)

# Count missing values in each column
print("Missing Data Counts:")
print(df.isna().sum())
```

```python
# Basic Summary Statistics
print("Summary Statistics:")
print(df.describe())
```

```python
# Plotting Histograms
plt.figure(figsize=(10, 6))
df["ISO"].plot(kind="hist", bins=50, title="ISO Distribution")
plt.xlabel("ISO")
plt.show()

plt.figure(figsize=(10, 6))
df["FocalLength"].plot(kind="hist", bins=20, title="Focal Length Distribution")
plt.xlabel("Focal Length (mm)")
plt.show()

plt.figure(figsize=(10, 6))
df["FNumber"].plot(kind="hist", bins=20, title="Aperture (F-Stop) Distribution")
plt.xlabel("F-Stop")
plt.show()

plt.figure(figsize=(10, 6))
df["ShutterSpeed"].plot(kind="hist", bins=20, title="Shutter Speed Distribution")
plt.xlabel("Shutter Speed (1/x sec)")
plt.show()
```

```python
# Correlation Matrix
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python

# Scatter Plot of Focal Length vs F-Number
plt.figure(figsize=(10, 6))
plt.scatter(df["FocalLength"], df["FNumber"], alpha=0.7)
plt.title("Focal Length vs. Aperture (F-Stop)")
plt.xlabel("Focal Length (mm)")
plt.ylabel("F-Stop")
plt.grid(True)
plt.show()
```

```python

# Boxplot of ISO vs Focal Length
plt.figure(figsize=(10, 6))
df.boxplot(column="ISO", by="FocalLength", grid=False)
plt.title("ISO Distribution by Focal Length")
plt.suptitle("")  # Remove default subtitle
plt.xlabel("Focal Length (mm)")
plt.ylabel("ISO")
plt.xticks(rotation=90)
plt.show()
```
